```python
import numpy as np
import pandas as pd
```

```python
from apyori import apriori
```

```python
reviews = pd.read_csv('../modified-bgg-15m-reviews.csv')
```

```python
reviews.head(10)
```

```python
reviews_lists = reviews.groupby('user').name.apply(list)
```

```python
reviews_lists
```

```python
# put the user for whom you want recommendations
test_user = 'Quinns'
```

```python
great = []
good = []
bad = []
games = []
user_games = reviews.loc[reviews['user'] == test_user]

for i in range(len(user_games)):
    game = user_games.iloc[i]['name']
    rating = user_games.iloc[i]['rating']
    games.append(game)
    if rating > 8:
        great.append(game)
    elif rating > 6:
        good.append(game)
    else:
        bad.append(game)
   

print(great)
print(good)
print(bad)
```

```python
users = reviews_lists.index.tolist()
users_with_great = []
for user in users:
    if len(set(great) & set(reviews_lists.loc[user])) > 3:
        users_with_great.append(user)

users_with_great
```

```python
great_df = reviews.loc[reviews['rating'] >= 9.0]
print(great_df.shape)
great_df = great_df.loc[great_df['user'].isin(users_with_great)]
# nines = reviews.loc[reviews['rating'] == 9.0]
# eights = reviews.loc[reviews['rating'] == 8.0]
# sevens = reviews.loc[reviews['rating'] == 7.0]
# sixes = reviews.loc[reviews['rating'] == 6.0]
# fives = reviews.loc[reviews['rating'] == 5.0]
# fours = reviews.loc[reviews['rating'] == 4.0]
# threes = reviews.loc[reviews['rating'] == 3.0]
# twos = reviews.loc[reviews['rating'] == 2.0]
# ones = reviews.loc[reviews['rating'] == 1.0]
```

```python
great_df.shape
```

```python
great_lists = great_df.groupby('user').name.apply(list)
```

```python
len(great_lists)
```

```python
# experimentally I found that 0.0057143 is a good balance of efficiency and utility; 
# if it's a support count of less than 150 or more than 400, though, it's changed to be in that 
# support count range.  Again, experimentally I found that this range is appropriate for generating useful results

support = 0.0057143
if 1 - (len(great_lists) - 400)/len(great_lists) < support:
    support = 1 - (len(great_lists) - 400)/len(great_lists)
elif 1 - (len(great_lists) - 150)/len(great_lists) > support:
    support = 1 - (len(great_lists) - 150)/len(great_lists)

association_rules = apriori(great_lists, min_support = support, min_confidence = 0.5)
association_results = list(association_rules)
association_results
```

```python
relevant_rules = []
for result in association_results:
    if len(set(great) & set(result[0])) > 1 and not all(game in games for game in result[0]):
        relevant_rules.append(result)

relevant_rules
```

```python
points = {}
recommenders = {}

for rule in relevant_rules:
    for game in rule[0]:
        points[game] = 0
        recommenders[game] = []
          
# these restrictions prevent double counting C for [A, B, C, D] and [A, B, C, E]
for rule in relevant_rules:
    for game in rule[0]:
        shared = set(rule[0]) & set(great)
        if not len(set(recommenders[game]) & shared) == len(shared):
            if not game in games:
                recommenders[game].extend(shared.difference(shared & set(recommenders[game])))
                points[game] = points[game] + 1
            
dict(sorted(points.items(), key=lambda item: item[1]))
            
```
